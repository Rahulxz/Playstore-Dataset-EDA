# **Checking Mean Rating for each Category of App**

```python
# Drop duplicates and assign the result back to a new variable or overwrite the original
play_no_duplicates = play.drop_duplicates()

# Remove '+' and ',' characters
play_no_duplicates['Installs'] = play_no_duplicates['Installs'].astype(str).str.replace('+', '', regex=False).str.replace(',', '', regex=False)
play_no_duplicates['Installs'] = pd.to_numeric(play_no_duplicates['Installs'], errors='coerce')

# To find the total installs per category and then sort, you'd sum the installs
new_play = play_no_duplicates.groupby("Category")["Installs"].sum().sort_values(ascending=False).reset_index()

# Get the top 10 categories by total installs
top_10 = new_play.head(10)
top_10
```

```python
plt.figure(figsize=(10,6))
sns.barplot(data=top_10,x='Category',y="Installs",hue='Category')
plt.xlabel("Category")
plt.ylabel("Total Installs")
plt.title("Top 10 Categories by Total Installs")
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

# **Corelational Heatmap Plot**
Checking out corelation between Installs, Ratings and Reviews.

```python
# Assuming 'play' and 'review' are your original dataframes
# and 'infomation' is the merged dataframe

# Ensure 'Installs' in the original 'play' dataframe is clean and numerical
play_cleaned = play.copy() # Work on a copy to avoid modifying the original 'play'
play_cleaned['Installs'] = play_cleaned['Installs'].astype(str).str.replace('+', '', regex=False).str.replace(',', '', regex=False)
play_cleaned['Installs'] = pd.to_numeric(play_cleaned['Installs'], errors='coerce') # Convert to numeric

# Perform the merge with the cleaned 'play' dataframe
infomation_cleaned_merge = pd.merge(play_cleaned, review, on='App')

# Drop rows with NaN in the relevant columns from the merged dataframe
infomation_for_heatmap = infomation_cleaned_merge.dropna(subset=['Rating', 'Reviews', 'Installs']).copy()

# Select the columns for the heatmap
selected_cols = infomation_for_heatmap[['Rating', 'Reviews', 'Installs']]

# Calculate the correlation matrix for the selected columns
correlation_matrix_selected = selected_cols.corr()

# Create the heatmap
plt.figure(figsize=(8, 6))
sns.heatmap(correlation_matrix_selected, annot=True, cmap='coolwarm', fmt=".2f")
plt.title("Correlation Heatmap of Rating, Reviews, and Installs")
plt.show()
```
